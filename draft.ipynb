```python
from nltk.corpus import stopwords
import nltk
import preProcessing
import globals as g
```

```python
data = []

for article in g.cleaned_text_train:
    data.append(nltk.word_tokenize(article))

DF = {}
for i in range(len(data)):
    tokens = data[i]
    for w in tokens:
        try:
            DF[w].add(i)
        except Exception:
            DF[w] = {i}
DF
```

```python
import math
DF_count = []
for i in DF:
    DF_count.append(len(DF[i]))

total_vocab = [x for x in DF]
num_of_words = len(total_vocab)

tf_idf = []
for i in range(len(data)):
    tfidf_doc = []
    n_of_words = len(data[i])
    words = data[i]
    for word in DF.items():
        # check the keys in dict. if it doesn't exist in the doc continue
        if i not in word[1]:
            tfidf_doc.append(0)
            continue
        else:
            n_of_occur = data[i].count(word[0])
            tfidf_doc.append((n_of_occur / n_of_words) * (math.log10(n_of_words / float(len(word[1])))))
    tf_idf.append(tfidf_doc)
    
```
